```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.head()
```

```python
df.tail()
```

```python
df.describe()
```

```python
# Relationship between Years and per CapitalIncome
plt.scatter(df['year'], df['per capita income (US$)'], color = 'lightcoral')
plt.title('Years vs Per Capital Income(US$)')
plt.xlabel('years')
plt.ylabel('per capita income(US$)')
#plt.box(False)
plt.show()
```

```python
X = df.iloc[:, :1]  # independent
y = df.iloc[:, 1:]  # dependent
```

```python
X
```

```python
train_test_split(X, y, test_size = 0.2, random_state = 40)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
plt.scatter(X,y)
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(X_train, y_train)
```

```python
Y_pred_test = reg.predict(X_test) 
Y_pred_test
```

```python
y_pred_train = reg.predict(X_train)
y_pred_train
```

```python
plt.scatter(X_train, y_train, color = 'lightcoral')
plt.plot(X_train, y_pred_train, color = 'firebrick')
plt.title('per Capital Income vs Years (Training Set)')
plt.xlabel('years')
plt.ylabel('per capita income(US$)')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Per Cap income/years', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
print(f'Coefficient: {reg.coef_}')
print(f'Intercept: {reg.intercept_}')
print('y = ', reg.intercept_, '+', reg.coef_, '* X')
```

```python
print('MSE = ', mean_squared_error(y_test, Y_pred_test ))
print('R2 = ', r2_score(y_test, Y_pred_test ))
print('RMSE = ', np.sqrt(mean_squared_error(y_test, Y_pred_test )))
```

```python
plt.scatter(X_train, y_train, color = 'lightcoral')
plt.plot(X_train, y_pred_train, marker= "o" ,color = 'firebrick')
plt.title('Per Capital Income(US$) vs Years (Training Set)')
plt.xlabel('year')
plt.ylabel('per capita income(US$)')
plt.legend(['X_train/y_train', 'X_train/Pred(y_test)'], title = 'Per Cap Income/Years', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
plt.scatter(X_test, y_test, color = 'lightcoral')
plt.plot(X_test, Y_pred_test, marker= "o" ,color = 'firebrick')
plt.title('Years vs Per Capital Income (US$) (Test Set)')
plt.xlabel('year')
plt.ylabel('per capita income(US$)')
plt.legend(['X_train/y_test', 'X_train/Pred(y_test)'], title = 'Per Cap income/years', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
Y_pred=reg.predict(X_test)
accuracy=r2_score(y_test,Y_pred)
print("the accuracy = ",int(accuracy*100),"%")
```
