```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
df = sns.load_dataset("tips")
```

```python
# Relationship between total bill and tips
plt.scatter(df['total_bill'], df['tip'], color = 'lightcoral')
plt.title('Total bill vs tips')
plt.xlabel('total_bill')
plt.ylabel('tip')
#plt.box(False)
plt.show()
```

```python
X = df.iloc[:, :1]  # independent
y = df.iloc[:, 1:2]  # dependent
```

```python
X
```

```python
y
```

```python
train_test_split(X, y, test_size = 0.2, random_state = 40)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
plt.scatter(X,y)
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_train, y_train)
```

```python
y_pred_test = reg.predict(X_test) 
y_pred_test
```

```python
y_pred_train = reg.predict(X_train)
y_pred_train
```

```python
plt.scatter(X_train, y_train, color = 'lightcoral')
plt.plot(X_train, y_pred_train, color = 'firebrick')
plt.title('Total Bill vs Tips (Training Set)')
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.legend(['X_train/)y_train', 'X_train/Pred(y_test)'], title = 'total bill/tip', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
reg.predict([[10.34]])
```

```python
print(f'Coefficient: {reg.coef_}')
print(f'Intercept: {reg.intercept_}')
print('y = ', reg.intercept_, '+', reg.coef_, '* X')
```

```python
print('MSE = ', mean_squared_error(y_test, y_pred_test ))
print('R2 = ', r2_score(y_test, y_pred_test ))
print('RMSE = ', np.sqrt(mean_squared_error(y_test, y_pred_test )))
```
